### List comprehension (포함형 / 지능형 / 축약형 리스트)
- 하나의 리스트에 그것 포함한 다른 리스트를 생성함
- list + for loop 개념
- [A for B in C] : C에 있는 B에서 A를 가져와라 => 리스트에 담아서 반환
    - A : 내가 원하는 출력 형태
    - B(A') : C의 형태
    - C : 반복 가능한 자료형 (리스트)
    
- 조건 추가 : [A for A' in C if ...]

```python
[color[0] for color in colors]   #0번째 인덱스만 반환
```

### NLTK
- Frequency Distribution : 단어가 몇 번 출현했는지 (빈도수)
- 텍스트의 성질 알기위해
    - 단어 개수
    - 단어 빈도수
- FreqDist(text)

```python
import nltk
from nltk.book import *
nltk.download('book', quiet=True)
```

```python
fdist = FreqDist(text1)
fdist
```

```python
#most_commom 함수 :제일 빈번한 상위 ()개 뽑아줌
fdist.most_common(50)
```

```python
fdist['that'], text1.count('that')  #여기서는 가능
```

```python
#비율로도 알 수 있다
fdist.freq('that')
```

```python
#그림도 가능
fdist.plot(50, cumulative=True)
```

```python
# 가장 빈번하지 않은 것 => 1번만 나타나는 단어들
fdist.hapaxes()

 #알파벳만 리스트에 담아보자
set([word for word in fdist.hapaxes() if word.isalpha()])
```

```python
# 대소문자 구분 없이 freq dist 만드려면? the 15000, The 600 => the 21000
fdist2 = FreqDist([word.lower() for word in text1])
fdist2['the']
```

```python
#다시 설명해주심
fdist = FreeDist(text1)  #대소문자, 첫글자 대문자, THE
fdist2 = FreeDist([word.lower() for word in text1]) #한번에 세어질 수 있도록
```

```python
# text2에서 길이가 3자 이하이고, 빈도수가 5이상
#내가 푼것
fdist3 = FreeDist([word for word in text2 if len(word)<=3 and fdist[word]>=5])

```

### 리스트 <-> 스트링
- split() : string.split(seperator)
    - default = ' '
- join() : seperator.join(sring)

```python
sent = 'I, have, a, car'
#tokens ['I', 'have', 'a', 'car'] 만들기 -> split사용하기
sent.split('.')
```

```python
# solution 
sent_tokens = [word.strip() for word in sent.split(',')]
sent_tokens

```

```python
#다시 한 문장으로 붙이기 - join
'-'.join(sent_tokens)
```

```python
#2
import nltk
from nltk. book import * 
nltk.download('book', quiet=True)

original = set([word for word in text2 if word.endswith('ch')])  #중복 X
word2 = []   
for word in original:
    if word.endswith('ch'):
        word2.append(word[0] + '-'*len(word[1:-2]) + 'ch')
word2.sort(key = len)
print(word2)
```
